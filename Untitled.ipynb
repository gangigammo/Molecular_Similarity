```python
from rdkit import rdBase, Chem
from rdkit.Chem import Draw, MACCSkeys, AllChem
from rdkit import DataStructs
from rdkit.Chem import MACCSkeys
from rdkit.Chem import AllChem

suppl = Chem.SDMolSupplier('./bmcmpd.sdf')
SDMols = [x for x in suppl if x is not None]
MACCS = [MACCSkeys.GenMACCSKeys(x) for x in SDMols]
ECFP4 = [AllChem.GetMorganFingerprintAsBitVect(x,2) for x in SDMols]
pair_list = []
MAXSIZE = 3
for i in range(MAXSIZE):
    pair_list.append([-1,-1,-1])
for i in range(20000):
    for j in range(i+1,20000):
        MACCS_sim = DataStructs.FingerprintSimilarity(MACCS[i],MACCS[j])
        ECFP4_sim = DataStructs.FingerprintSimilarity(ECFP4[i],ECFP4[j])
        if  ECFP4_sim - MACCS_sim > pair_list[MAXSIZE-1][0]:
            pair_list[MAXSIZE-1][0] = ECFP4_sim - MACCS_sim
            pair_list[MAXSIZE-1][1] = i
            pair_list[MAXSIZE-1][2] = j
            t1 = 2
            while MAXSIZE - t1 >= 0 and  pair_list[MAXSIZE - t1 + 1][0] > pair_list[MAXSIZE - t1][0] :
                pair_list[MAXSIZE - t1],pair_list[MAXSIZE - t1 + 1] = pair_list[MAXSIZE - t1 + 1] ,pair_list[MAXSIZE - t1]
                t1 = t1+1;

print(pair_list)
for i in range(3):
    print(DataStructs.FingerprintSimilarity(MACCS[pair_list[i][1]],MACCS[pair_list[i][2]]))
    print(DataStructs.FingerprintSimilarity(ECFP4[pair_list[i][1]],ECFP4[pair_list[i][2]]))
Draw.MolToFile(SDMols[pair_list[0][1]],'images/mol1_1.o.png')
Draw.MolToFile(SDMols[pair_list[0][2]],'images/mol1_2.o.png')
Draw.MolToFile(SDMols[pair_list[1][1]],'images/mol2_1.o.png')
Draw.MolToFile(SDMols[pair_list[1][2]],'images/mol2_2.o.png')
Draw.MolToFile(SDMols[pair_list[2][1]],'images/mol3_1.o.png')
Draw.MolToFile(SDMols[pair_list[2][2]],'images/mol3_2.o.png')
```
